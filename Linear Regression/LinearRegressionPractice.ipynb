```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

In this I will be completing the practice problem at the end of the second video

The goal is to find Canada's estimated per capita income for the year 2020

```python
df = pd.read_csv('CanadaPerCapitaIncome.csv')
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.perCapitaIncome)
```

```python
%matplotlib inline

plt.title('Canadian Per Capita Income since 1970')
plt.ylabel('Year')
plt.xlabel('Per Capita Income(USD)')
plt.scatter(df.year, df.perCapitaIncome, color = 'red')
plt.plot(df.year, reg.predict(df[['year']]), color = 'blue')
```

```python
reg.predict([[2020]])
```
